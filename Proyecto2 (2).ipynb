# Pruebas de hipotesis

# **Variable 1: Desempleo**

**H0:** No hay diferencia significativa en las tasas de desempleo entre Colombia y Suramérica
**H1:** Existe una diferencia significativa en las tasas de desempleo entre Colombia y suramerica

```python
for year in df['año'].unique():
  print (year)

  df3=df.query('año==@year')

  colombia = df3[df3['pais'] == 'Colombia']
  suramerica = df3[df3['pais'].isin(['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela, RB'])]

  desempleo_colombia = colombia['desempleo']
  desempleo_suramerica = suramerica['desempleo']

  t_statistic, p_value = stats.ttest_1samp(desempleo_suramerica, desempleo_colombia)

  alpha = 0.05

  print(f'Valor de la estadística t: {t_statistic}')
  print(f'Valor p: {p_value}')

  if p_value < alpha:
      print('Rechazamos la hipótesis nula (H0): Existe una diferencia significativa en las tasas de desempleo entre Colombia y Suramérica.')
  else:
      print('No podemos rechazar la hipótesis nula (H0): No hay evidencia suficiente para afirmar una diferencia significativa en las tasas de desempleo entre Colombia y Suramérica.')
```

**H0:** No hay diferencia significativa en las tasas de desempleo entre Colombia y Norteamérica
**H1:** Existe una diferencia significativa en las tasas de desempleo entre Colombia y Norteamérica

```python
for year in df['año'].unique():
  print (year)

  df3=df.query('año==@year')

  colombia = df3[df3['pais'] == 'Colombia']
  norteamerica = df3[df3['pais'].isin(['Estados Unidos', 'Canada', 'Mexico', 'Costa Rica'])]

  desempleo_colombia = colombia['desempleo']
  desempleo_norteamerica = norteamerica['desempleo']

  t_statistic, p_value = stats.ttest_1samp(desempleo_norteamerica, desempleo_colombia)

  alpha = 0.05

  print(f'Valor de la estadística t: {t_statistic}')
  print(f'Valor p: {p_value}')

  if p_value < alpha:
      print('Rechazamos la hipótesis nula (H0): Existe una diferencia significativa en las tasas de desempleo entre Colombia y Norteamérica.')
  else:
      print('No podemos rechazar la hipótesis nula (H0): No hay evidencia suficiente para afirmar una diferencia significativa en las tasas de desempleo entre Colombia y Norteamérica.')
```

# **Variable 2: PIB (Producto Interno Bruto)**

H0: No hay diferencia significativa en el PIB entre Colombia y el país de comparación y Suramérica
H1: Existe una diferencia significativa en el PIB entre Colombia y Suramerica.

```python
for year in df['año'].unique():
  print (year)

  df3=df.query('año==@year')

  colombia = df3[df3['pais'] == 'Colombia']
  suramerica = df3[df3['pais'].isin(['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Paraguay', 'Peru', 'Uruguay', 'Venezuela, RB'])]

  pib_colombia = colombia['pib']
  pib_suramerica = suramerica['pib']

  t_statistic, p_value = stats.ttest_1samp(pib_suramerica, pib_colombia)

  alpha = 0.05

  print(f'Valor de la estadística t: {t_statistic}')
  print(f'Valor p: {p_value}')

  if p_value < alpha:
      print('Rechazamos la hipótesis nula (H0): Existe una diferencia significativa en el Ingreso Per Cápita entre Colombia y Suramérica.')
  else:
      print('No podemos rechazar la hipótesis nula (H0): No hay evidencia suficiente para afirmar una diferencia significativa en el Ingreso Per Cápita entre Colombia y Suramérica.')
```

**H1:** Existe una diferencia significativa en el Ingreso Per Cápita entre Colombia y Norteamérica

```python
for year in df['año'].unique():
  print (year)

  df3=df.query('año==@year')

  colombia = df3[df3['pais'] == 'Colombia']
  norteamerica = df3[df3['pais'].isin(['Estados Unidos', 'Canada', 'Mexico', 'Costa Rica'])]

  pib_colombia = colombia['pib']
  pib_norteamerica = norteamerica['pib']

  t_statistic, p_value = stats.ttest_1samp(pib_norteamerica, pib_colombia)

  alpha = 0.05

  print(f'Valor de la estadística t: {t_statistic}')
  print(f'Valor p: {p_value}')

  if p_value < alpha:
      print('Rechazamos la hipótesis nula (H0): Existe una diferencia significativa en el Ingreso Per Cápita entre Colombia y Norteamérica.')
  else:
      print('No podemos rechazar la hipótesis nula (H0): No hay evidencia suficiente para afirmar una diferencia significativa en el Ingreso Per Cápita entre Colombia y Norteamérica.')
```
